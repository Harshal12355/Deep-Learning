```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.random.uniform(0, 1, 100)
```

```python
def stochastic_gradient_descent(x, y, lr, iterations, y_star=np.mean(x)):
    MSE = []

    for iteration in range(iterations):
        i = np.random.randint(0, len(x)) # choosing a random integer from x
        gradient = 2*(y - x[i])
        y = y - (lr * gradient)
        mse = (y_star - y)**2
        MSE.append(mse)
    plt.plot(MSE, label=f"Learning rate: {lr}")
    plt.xlabel("Iteration")
    plt.ylabel("MSE Error")
    plt.title("MSE Error vs Iterations for SGD")
    plt.legend()
    plt.show()
```

```python
stochastic_gradient_descent(x, y=0, lr=0.01, iterations=1000)
```

```python
stochastic_gradient_descent(x, y=0, lr=0.001, iterations=1000)
```

```python
stochastic_gradient_descent(x, y=0, lr=0.1, iterations=1000)
```
